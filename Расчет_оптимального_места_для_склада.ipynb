```
import pandas as pd
import numpy as np
from scipy.spatial import distance
```

```
# Загружаем данные из файла

df = pd.read_csv('/content/Копия datasets address - datasets.csv')
df.info()
```

```
df[df['latitude'] == '59.829495_'] = 59.829495
df[df['latitude'] == '59.742824.'] = 59.742824
```

```
df[df['latitude'] == '59.829495_']
```

```
df['latitude'] = df['latitude'].astype(float)
df.head()
```

```
df['longitude'] = df['longitude'].astype(float)
df.head()
```

```
df[df['shipments'] == '2,00']
```

```
df['shipments'] = df['shipments'].replace('2,00', 2)
df['shipments'] = df['shipments'].replace('4,00', 4)
df['shipments'] = df['shipments'].replace('1,00', 1)
df['shipments'] = df['shipments'].replace('3,00', 3)
df = df.dropna()
```

```
df['shipments'] = df['shipments'].astype(int)
```

```
df.head()
```

```
latitude = np.array(df['latitude'])
longitude = np.array(df['longitude'])
shipments = np.array(df['shipments'])
village = df['village'].to_list()
```

```
data = pd.DataFrame({'Долгота, км': longitude,
                     'Широта, км': latitude, 
                     'Доставки': shipments}, index=village)
```

```
data.head()
```

```
vectors = data[["Долгота, км", "Широта, км"]].values
distances = []
for village_from in range(len(village)):
    row = []
    for village_to in range(len(village)):
        value = distance.euclidean(vectors[village_from], vectors[village_to])
        row.append(value)
    distances.append(row)
 
shipping_in_week = []
for i in  range(len(village)):
    shipping_in_week.append(2 * np.dot(np.array(distances[i]), shipments))
 
shipping_in_week_df = pd.DataFrame({'Расстояние': shipping_in_week}, index=village)
 
print(shipping_in_week_df)
 
print()
print('Населённый пункт для склада:', 
      shipping_in_week_df['Расстояние'].idxmin())
```
